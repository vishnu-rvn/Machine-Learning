```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = sns.load_dataset('iris')
```

```python
df['species'].unique()
```

```python
def convert(species):
    if species == 'setosa':
        return 0
    elif species == 'versicolor':
        return 1
    elif species == 'virginica':
        return 2
```

```python
df.head()
```

```python
df['species_num'] = df['species'].map(convert)
```

```python
df.head()
```

```python
clust = df.drop(['species','species_num'], axis = 1)
```

```python
clust.head()
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(clust)
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1,ncols=2, figsize = (12,6))
ax1.set_title('Original')
ax1.scatter(df['petal_width'],df['petal_length'],c = df['species_num'])
ax2.set_title('KMeans')
ax2.scatter(df['petal_width'],df['petal_length'],c = kmeans.labels_)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1,ncols=2, figsize = (12,6))
ax1.set_title('Original')
ax1.scatter(df['sepal_width'],df['sepal_length'],c = df['species_num'])
ax2.set_title('KMeans')
ax2.scatter(df['sepal_width'],df['sepal_length'],c = kmeans.labels_)
```
